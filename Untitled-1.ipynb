```python
import pandas as pd
import re

df_first_page = pd.read_csv(r"C:\Users\coys7\Desktop\krisha\app-s_first_page")

#splitiing main info into different columns
df_first_page[['Rooms', 'Square_in_m^2', 'Floor']]=df_first_page['main_info'].str.split(",", expand=True)

#cleaning data from unnecessary info
df_first_page['prices']=df_first_page['prices'].str.replace('[\n ]', '', regex=True)
df_first_page['address']=df_first_page['address'].str.replace('[\n]', '', regex=True)
df_first_page['description']=df_first_page['description'].str.replace('[\n]', '', regex=True)
df_first_page['Rooms']=df_first_page['Rooms'].str.replace('-комнатная квартира', '', regex=True)
df_first_page['Square_in_m^2']=df_first_page['Square_in_m^2'].str.replace(' м²', '', regex=True)
df_first_page['Floor']=df_first_page['Floor'].str.replace(' этаж', '', regex=True)

#working with description
df1=df_first_page['description']#.str.split(',', expand=True)
df1=pd.DataFrame(df1)

def extract_info(text):
    pattern = r'[а-яА-ЯёЁ]+ дом'
    find_pattern = re.findall(pattern, text)
    return find_pattern

df1['Type'] = df1['description'].apply(lambda x:extract_info(x))


#def count(type):
#    count_1=len(type)
#    while count_1 > 1:
#        del type(1)
#    return count_1
#df1['count'] = df1['Type'].apply(lambda x:count(x))
def split_Type(arr):
    length=len(arr)
    while length > 1:
        split_arr1 = arr[len(arr)//2:]
    return split_arr1
df1['trash'] = df1['Type'].apply(lambda x:split_Type(x))
df1
```
